```python
import csv
import glob
import os

input_path = r'/content/new/'
output_file = r'/content/new/hanspell_sen+str.csv'

file_list = glob.glob(input_path + '*.csv')

with open(output_file, 'w', encoding='utf-8-sig') as f:
    for i, file in enumerate(file_list):
        with open(file, 'r', encoding='utf-8-sig') as f2:
            while True:
                line = f2.readline()
                if not line:
                    break
                f.write(line)
        print(file.split('\\')[-1])

    else:
        with open(file, 'r', encoding='utf-8-sig') as f2:
            n = 0
            while True:
                line = f2.readline()
                if n != 0:
                    f.write(line)
                if not line:
                    break
                n + 1
        print(file.split('\\')[-1])

file_num = len(next(os.walk('/content/new/'))[2])
print(file_num, 'file merge complete...')
```

```python
from konlpy.tag import Kkma, Komoran, Okt, Mecab
import pandas as pd
import csv
import sys
import re
import datetime as dt
import numpy as np
import re

kkm = Kkma()
file_path = './total_story_pre_0803.txt'
df = open(file_path, mode='r', encoding='utf-8')
tdata = df.read()  # 파일 읽기
print(type(tdata))
tdata = tdata.split('/////')
tdata = list(filter(None, tdata))
print(len(tdata))


number = [0, 100, 200, 300, 400, 500, 600, 700, 800, 900, 1000, 1100, 1200, 1306]
for i in range(len(number)):
  if i == len(number)-1 :
    break
  else:
    start = number[i]
    end = number[i+1]

    # 동화 끊기
    data = tdata[start:end]

    story = []
    sentence = []

    a = 0

    for d in data:
      a += 1
      if a % 10 == 0:
        print(a)
      if len(d) > 1:
        d = kkm.sentences(d)  # 문장단위로 추출됨
        st = ".".join(d)  # 문장단위 .으로 구분하여 하나의 동화 생성
        st = st.replace('..', '.')
        # print(st)
        story.append(st)
        for l in d:  # kkm처리한 story의 문장 추출
          sentence.append(l)

    # print(sentence)
    # print(story)

    str_prompt = []
    str_completion = []

    for s in story:
      s = re.split(r'[."]', s)
      for j in range(len(s)):
        if j % 2 == 0:
          prompt = s[j:j + 3]
          comple = s[j + 3: j + 6]
          prompt = " ".join(prompt)
          comple = " ".join(comple)
          if len(prompt) > 0 and len(comple) > 0:
            prompt = re.sub(r'[^가-힣 .]', '', prompt)
            comple = re.sub(r'[^가-힣 .]', '', comple)
            str_prompt.append(prompt)
            str_completion.append(comple)

    sen_prompt = []
    sen_completion = []

    for se in sentence:  # 한 문장씩 추출
      try:
        se = re.split(r'[? ,!.]', str(se))
        # print(se)
        prompt = se[:3]
        prompt = " ".join(prompt)
        comple = se[3:]
        comple = " ".join(comple)

        if len(prompt) > 0 and len(comple) > 0:
          prompt = re.sub(r'[^가-힣 .]', '', prompt)
          comple = re.sub(r'[^가-힣 .]', '', comple)
          sen_prompt.append(prompt)
          sen_completion.append(comple)
      except:
        pass

    prompt = str_prompt + sen_prompt
    completion = str_completion + sen_completion

    ldata = {'prompt': prompt, 'completion': completion}

    df = pd.DataFrame(ldata, columns=['prompt', 'completion'])
    del_index = df[(df['prompt'] == ' ') | (df['completion'] == ' ')].index
    df.drop(del_index, inplace=True)
    df.reset_index(inplace=True, drop=True)
    df.to_csv(f'{dt.date.today()}pre_sen_{end}.csv', encoding='utf-8-sig')

  print(f'{end}완료')
```

```python
from hanspell import spell_checker
import pandas as pd
import csv
import sys
import re
import datetime as dt
import numpy as np
import re

file_path = './total_story_pre_0803.txt'
df = open(file_path, mode='r', encoding='utf-8')
tdata = df.read()  # 파일 읽기
print(type(tdata))
tdata = re.compile(r'[^가-힣!."?/]').sub('',tdata)
tdata = tdata.split('/////')
tdata = list(filter(None, tdata))
print(len(tdata))


number = [800, 900, 1000, 1100, 1200, 1306]
for i in range(len(number)):
  if i == len(number)-1 :
    break
  else:
    start = number[i]
    end = number[i+1]

    # 동화 끊기
    data = tdata[start:end]

    story = []
    sentence = []

    a = 0

# hanspell 처리 부분
    story = []
    for s in data:
      a += 1
      if a % 10 ==0:
        print(a, '완료')
      s = s.split('.')        #s = re.split(r'[."]', s)
      total_sentence = []
      try:
        for i in range(len(s)):
          spell_sentence = spell_checker.check(s[i]) #처리
          spell_sentence = spell_sentence.checked
          if len(spell_sentence) > 2:
            total_sentence.append(spell_sentence + '.')
        total_sentence = ' '.join(total_sentence)  
        print(total_sentence)
        story.append(total_sentence)
      except:
        continue

#문장 생성 부분
    str_prompt = []
    str_completion = []

    for s in story:
      s = re.split('([."?!])', s) #띄어쓰기X
      for j in range(len(s)):
        if j % 2 == 0:
          prompt = s[j:j + 2]
          comple = s[j + 2: j + 6]
          prompt = "".join(prompt)
          comple = "".join(comple)
          prompt = re.sub(r'[^가-힣 ,.?!]', '', prompt)
          comple = re.sub(r'[^가-힣 ,.?!]', '', comple)
          if len(prompt) > 0 and len(comple) > 0:
            str_prompt.append(prompt)
            str_completion.append(comple)  


#어절 생성 부분
    sen_prompt = []
    sen_completion = []

    for s in story:
      try:
        s = re.split(r'[." ?!]', s) #띄어쓰기 있
        for j in range(len(s)):
          if j % 4 == 0:
            prompt = s[j:j + 4]
            comple = s[j + 4: j + 16]
            prompt = " ".join(prompt)
            comple = " ".join(comple)
            prompt = re.sub(r'[^가-힣 .]', '', prompt)
            comple = re.sub(r'[^가-힣 .]', '', comple)
            if len(prompt) > 5 and len(comple) > 5:
              sen_prompt.append(prompt)
              sen_completion.append(comple)
      except:
        pass        

    prompt = sen_prompt + str_prompt
    completion = sen_completion + str_completion

    ldata = {'prompt': prompt, 'completion': completion}

    df = pd.DataFrame(ldata, columns=['prompt', 'completion'])
    del_index = df[(df['prompt'] == ' ') | (df['completion'] == ' ')].index
    df.drop(del_index, inplace=True)
    df.reset_index(inplace=True, drop=True)
    df.to_csv(f'{dt.date.today()}_hanspell_sen_{end}.csv', encoding='utf-8-sig')

  print(f'{end}완료')
```
